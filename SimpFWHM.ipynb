```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as spsig
```

```python
sigma = 2.

x = np.linspace(-10., 10., 1000)
y = np.exp( -(x/sigma)**2. / 2. ) / sigma / np.sqrt(2. * np.pi) + 0.1
```

```python
np.max(y)/2
```

```python
HMInd = np.argsort(np.abs(y - np.max(y)/2))[:2]
```

```python
HM_y = np.max(y)/2

plt.figure(dpi=500)
plt.plot(x, y)
plt.hlines(HM_y, x[HMInd[0]], x[HMInd[1]], colors='C1')
plt.text(np.mean(x[HMInd]), HM_y, f'{x[HMInd[1]] - x[HMInd[0]]:.2f}', ha='center', va='top')
plt.show()
```
